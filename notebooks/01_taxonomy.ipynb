This is just a custom script to get the full lineage using GTDB, ignore this if you already have one.
df_gtdb_meta is generated by BGCFlow by using genus name in the samples.csv

```python
import pandas as pd
```

```python
df_gtdb = pd.read_csv("df_gtdb_meta.csv", index_col="genome_id")
df_gtdb
```

```python
df_samples = pd.read_csv("../samples.csv", index_col="genome_id")
df_samples
```

```python
df_custom_taxonomy = pd.DataFrame()
for num, i in enumerate(df_gtdb.index):
    classification = []
    for c in ["Domain","Phylum","Class","Order","Family","Genus"]:
        classification.append(df_gtdb.loc[i, c])
    classification.append(f's__{df_samples.loc[i, "organism"]}')
    classification = ";".join(classification)
    df_custom_taxonomy.loc[num, "user_genome"] = i
    df_custom_taxonomy.loc[num, "classification"] = classification
df_custom_taxonomy.to_csv("../custom_taxonomy.tsv", index=False, sep="\t")
```
