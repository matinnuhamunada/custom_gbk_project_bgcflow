```python
import pandas as pd
from pathlib import Path
from Bio import SeqIO
import shutil

def extract_taxonomy(genbank_file_path, keyword="organism"):
    with open(genbank_file_path, "r") as handle:
        record = SeqIO.read(handle, "genbank")
    taxonomy = record.annotations.get(keyword, [])
    return taxonomy
```

```python
%%bash
rm -rf ../input_files
mkdir ../input_files
wget -P ../input_files https://github.com/NBChub/bgcflow/files/13499395/NW_003318762.1_0_43680_changed_species.txt
wget -P ../input_files https://github.com/NBChub/bgcflow/files/13499396/NW_019110264.1_0_31676_changed_species.txt
wget -P ../input_files https://github.com/NBChub/bgcflow/files/13499397/NW_003385442.1_0_46725_changed_species.txt
wget -P ../input_files https://github.com/NBChub/bgcflow/files/13499398/NZ_KE386655.1_0_25886_changed_species.txt
```

```python
data_path = Path("../input_files/")
input_files = [i for i in data_path.glob("*_changed_species.txt")]
```

```python
samples_csv = pd.DataFrame()
for num, i in enumerate(input_files):
    shutil.copy(i, i.parent / f"{i.stem}.gbk")
    samples_csv.loc[num, "genome_id"] = str(i.stem).strip("_changed_species")
    samples_csv.loc[num, "source"] = "custom"

    organism = extract_taxonomy(i)
    samples_csv.loc[num, "genus"] = organism.split()[0]
    samples_csv.loc[num, "species"] = organism
    samples_csv.loc[num, "organism"] = organism

    samples_csv.loc[num, "input_file"] = f"{i.stem}.gbk"

samples_csv.to_csv("../samples.csv", index=False)
samples_csv
```
